```python
#Oridnal Encoding

import pandas as pd
import numpy as np
data = ({'Subjects': ['BEE', 'MA', 'ME'],
        'Societies': ['AIMS','RT','CM'],
        'PO' : [10,12,25]
       })
df = pd.DataFrame(data)
print(df)
column_names = df.columns.tolist() #To convert columns to list
print(column_names)
print(df)

for col in column_names:
    if df[col].dtype == object:
        print("object_type")
    else:
        print ("not_object_type")

def ordinal_encoding(df):
    df['Subjects'] = df['Subjects'].map({'BEE': 1, 'MA': 2, 'ME': 3, 'AEC': 4})
    df['Societies'] = df['Societies'].map({'AIMS': 1, 'RT': 2, 'CM': 3})
    return df
df = ordinal_encoding(df)
print(df)
```
