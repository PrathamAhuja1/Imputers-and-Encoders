```python
#One Hot Encoding

import pandas as pd
import numpy as np

data = {'Subjects': ['BEE', 'MA', 'ME', 'AEC', 'Physics'],
        'Societies': ['AIMS','RT','CM','ECell','DCG']}

data_dataframe = pd.DataFrame(data)
print("Original DataFrame:")
print(data_dataframe)

def One_Hot_Encoding(data, column):
    unique_values = data[column].unique()
    new_columns = {}

    for values in unique_values:
        new_column = f"{column}_{values}"
        data[new_column] = np.where(data[column] == values, 1, 0)
        new_columns[values] = new_column

    return data.drop(column, axis=1), new_columns

encoded_dataframe, new_columns = One_Hot_Encoding(data_dataframe, 'Subjects')
encoded_dataframe_1, new_columns_1 = One_Hot_Encoding(encoded_dataframe, 'Societies')

print("\nDataFrame with Custom new_values:")
print(encoded_dataframe_1)

print("\nNew Columns:")
print(new_columns,new_columns_1)
```
