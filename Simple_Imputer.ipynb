```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame({"Value_1":[None, 224, 35, 29, 1,195, 74, 65, None, 19],
                   "Value_2":[None, None, None, 3,None, 2,None, 3, None,34],
                   "Value_3":[20, 9, None, 3, 8,None, 1, None, 8, None],
                   "Value_4":[10, None, 10, 10, None,10, 10, None, None, 10],
                   "Value_5":[None, None, None,None, None,None,None, None, 28, 28]})
print(data)
```

```python
def mean_imputer(data):
    data = data.convert_dtypes()
    column_names = data.columns.tolist()
    for i in column_names:
        data[i].fillna(int(data[i].mean()),inplace=True)
    return data


def median_imputer(data):
    data = data.convert_dtypes()
    column_names = data.columns.tolist()
    for i in column_names:
        data[i].fillna(int(data[i].median()),inplace=True)
    return data


def mode_imputer(df):
    data = data.convert_dtypes()
    column_names = data.columns.tolist()
    for i in column_names:
        data[i].fillna(int(max(data[i].mode())),inplace=True)
    return data


def Simple_Imputer(data,s ):
    if (s == "mean"):
        data =mean_imputer(data)
    elif (s=="median"):
         data =median_imputer(data)
    else:
        (s=="mode")
        data= mode_imputer(data)
    return data
```
